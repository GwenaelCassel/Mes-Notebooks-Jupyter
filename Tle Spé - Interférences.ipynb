## ETAPE 0 : Initialisation des bibliothèques python nécessaires

```python
import matplotlib.pyplot as plt #On charge la bibliothèque graphique
import numpy as np #On charge une bibliothèque de calculs
```

## ETAPE 1 : Choix de la valeur de l'amplitude A, de la période T et du déphasage  $\phi$ des ondes interférant entre-elles.
### $\color {red} {"A\: VOUS\: DE\: JOUER\,!"}$
### Essayer plusieurs valeurs de l'amplitude A et de la période T de l'onde. Observer les modifications sur la représentation temporelle en bas de page.

```python
A = 1 #L'amplitude A est fixée entre 0 et 10
T = 2.0 #La période est fixée entre 0 et 10
phi = 0 #Le déphasage est fixé entre 0 et 2pi
```

## ETAPE 2 : Création des deux ondes y1 et y2 interférant et de la somme y3 de ces deux ondes.

```python
t = np.linspace(0, 10, 5000) #On définit le nombre de points N d'abscisse t selon la durée de la période t, N = 5000/T, pour avoir un visuel toujours continu de la fonction

y1 = A*np.sin((2*np.pi/T)*t) #On calcule les ordonnées y1 des points de l'onde 1

y2 = A*np.sin((2*np.pi/T)*t+phi) #On calcule les ordonnées y2 des points de l'onde 2

y3 = y1 + y2 #On calcule la somme des ondes y1 et y2
```

## ETAPE 3 : Représentation de l'allure temporelle des deux ondes y1, y2 et de leur somme y3.

```python
Amax = max(np.max(y1), np.max(y3)) #Amax est le maximum d'amplitude entre y1, y2 et y3, il permet de définir l'amplitude verticale du graphique
plt.figure(figsize=(12,8)) #Définit la taille de l'affichage du graphique
plt.grid() #Affiche une grille sur le graphique
plt.axline((0,0),(1,0),color = "black") #Surligne l'axe des abscisses (horizontal) d'une droite noire
plt.title("Allure temporelle des ondes", size=18) #Définit le titre du graphique
plt.ylabel("Amplitude (m)", size=18) #Affiche le titre des ordonnées
plt.xlabel("Temps (s)", size=18) #Affiche le titre des abscisses
plt.xlim(0, 10) #Définit le début et la fin des abscisses sur le graphique
plt.ylim(-1.1*Amax, 1.7*Amax) #Définit le début et la fin des ordonnées sur le graphique
plt.plot(t,y1, label = 'y1', color ='blue') #Affiche les points expérimentaux
plt.plot(t,y2, label = 'y2', color ='green') #Affiche les points expérimentaux
plt.plot(t,y3,  label = 'y3', color ='red', linewidth = 4) #Affiche les points expérimentaux
plt.legend(loc = 'upper right', fontsize=18)
plt.text(2,1.4*Amax,"valeur du déphasage $\phi$ = " + str(phi), size=22, color='b', bbox=dict(boxstyle="round", ec=("k"), fc=("w"))) #Affiche l'équation de la droite dans un encadré
plt.show() #Affiche le graphique
```
