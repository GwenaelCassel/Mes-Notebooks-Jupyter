## Etape 0 : On initialise les bibliothèques utilisées dans ce programme

```python
import numpy as np #On importe la bibliothèque de calcul scientifique numpy
import matplotlib.pyplot as plt #On importe la bibliothèque pour les graphiques matplotlib
```

## Etape 1 : On précise les valeurs des différentes quantités nécessaires aux calculs
### $\color {red} {"A\: VOUS\: DE\: JOUER\,!"}$
### Modifier les valeurs de $V_A$ du volume de la solution d'acide, de la concentration $C_{B}$ de base, du volume $V_E$ versé à l'équivalence.

```python
VA = "A modifier" # (mL) Volume de solution titrée
CB = "A modifier" # (mol/L) Concentration de solution titrante
VE = "A modifier" # (mL) Volume de solution titrante versé à l'équivalence
CA = CB*VE/VA # (mol/L) Concentration initiale de solution titrée
```

## Etape 2 : On calcule l'évolution des quantités de matière de chaque espèce chimique

```python
# AVANT EQUIVALENCE
VB1 = np.linspace(0, VE, 20) #On génère les différentes valeurs de VB avant équivalence
nAH_1 = CA*VA*1E-3 - CB*VB1*1E-3 #Calcul de la quantité de l'acide AH, le facteur 1E-3 est la conversion du volume en L
nNa_1 = CB*VB1*1E-3 #Calcul de la quantité de l'ion Na+, le facteur 1E-3 est la conversion du volume en L
nOH_1 = 0*VB1 #On génère une liste de valeurs égales à 0 pour OH- avant l'équivalence
nA_1 = CB*VB1*1E-3 #Calcul de la quantité de base conjuguée A-, le facteur 1E-3 est la conversion du volume en L

# APRES EQUIVALENCE
VB2 = np.linspace(VE, 20, 20) #On génère les différentes valeurs de VB après équivalence
nAH_2 = 0*VB2 #On génère une liste de valeurs égales à 0 pour AH après l'équivalence
nNa_2 = CB*VB2*1E-3 #Calcul de la quantité de l'ion Na+, le facteur 1E-3 est la conversion du volume en L
nOH_2 = CB*(VB2-VE)*1E-3 #Calcul de la quantité de l'ion OH-, le facteur 1E-3 est la conversion du volume en L
nA_2 = 0*VB2 + CB*VE*1E-3 #On génère une liste pour la quantité de l'ion A-, le facteur 1E-3 est la conversion du volume en L
```

## Etape 3 : On représente graphiquement les évolutions de chaque espèce chimique

```python
plt.figure(figsize=(12,8)) #On précise les dimensions de la figure
plt.plot(VB1, nAH_1, "c-", label=r"$AH$") #La quantité de matière de AH avant l'équivalence est représentée en couleur cyan
plt.plot(VB2, nAH_2, "c-") #La quantité de matière de AH après l'équivalence est représentée en couleur cyan
plt.plot(VB1, nNa_1, "b-", label=r"$Na^+$") #La quantité de matière de Na+ avant l'équivalence est représentée de couleur bleue
plt.plot(VB2, nNa_2, "b-") #La quantité de matière de Na+ après l'équivalence est représentée de couleur bleue
plt.plot(VB1, nOH_1, "g-", label=r"$OH^-$") #La quantité de matière de OH- avant l'équivalence est représentée de couleur verte
plt.plot(VB2, nOH_2, "g-") #La quantité de matière de OH- après l'équivalence est représentée de couleur verte
plt.plot(VB1, nA_1, "r-", label=r"$A^-$" ) #La quantité de matière de A- avant l'équivalence est représentée de couleur rouge
plt.plot(VB2, nA_2, "r-") #La quantité de matière de A- après l'équivalence est représentée de couleur rouge
plt.axvline(VE, color="gray", linestyle ="dashed") #On génère une droite verticale en pointillés gris et d'abscisse VE
plt.xlabel("A modifier", size=18) #On génère le nom de l'axe des abscisses
plt.ylabel("A modifier", size=18) #On génère le nom de l'axe des ordonnées
plt.title("A modifier", size=18) #On génère un titre de police 18
plt.legend(fontsize=18) #On génère une légende de police 18
plt.text(VE-1, 0.75*np.max(nNa_2), 'VE = ' + str(VE) + ' mL', size=18, color="gray", bbox=dict(boxstyle="round", ec=("k"), fc=("w"))) #On génère une zone de texte grise qui précise la valeur de VE
plt.grid() #On génère une grille
plt.show() #On affiche les éléments du graphique
```
