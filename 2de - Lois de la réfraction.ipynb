### ETAPE 0 : Initialisation des bibliothèques python nécessaires.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.close() #On ferme les fenêtres graphiques éventuellement ouvertes.
```

### ETAPE 1 : Entrée des valeurs expérimentales de l'angle d'incidence "i" et de l'angle de réfraction "r"

### $\color {red} {"A\: VOUS\: DE\: JOUER\,!"}$ 
### Compléter les listes des valeurs d'angle d'incidence et d'angle de réfraction. Les nombres sont séparés par une virgule

```python
angle_i = np.array([]) #liste des valeurs des angles d'incidences
angle_r = np.array([]) #liste des valeurs des angles de réfraction
```

### ETAPE 2 : Construction du modèle de Grosseteste (12ème siècle)

```python
k_G = 2 #Valeur du coefficient directeur de la droite modèle

pas_G = 90/100 #pas générant les abscisses du modèle

X_modele_G = np.arange(0, 90, pas_G) #On génère les abscisses du modèle de 0 à 90
                                                        
for i in X_modele_G:
    Y_modele_G = k_G * X_modele_G #On génère les ordonnées du modèle
```

### ETAPE 3 : Construction du modèle de Kepler (16ème siècle - 17ème siècle)

```python
k_K = 1.5 #Valeur du coefficient directeur de la droite modèle

pas_K = 90/100 #pas générant les abscisses du modèle

X_modele_K = np.arange(0, 90, pas_K) #On génère les abscisses du modèle de 0 à 90
                                                        
for i in X_modele_K:
    Y_modele_K = k_K * X_modele_K #On génère les ordonnées du modèle
```

### ETAPE 4 : Construction du modèle de Descartes (17ème siècle)

```python
sin_i = np.sin(angle_i*3.1415/180) #Calcul des sinus des angles d'incidence
sin_r = np.sin(angle_r*3.1415/180) #Calcul des sinus des angles de réfraction
```

```python
k_D = 1.49 #Valeur du coefficient directeur de la droite modèle

pas_D = 1/100 #pas générant les abscisses du modèle

X_modele_D = np.arange(0, 1, pas_D) #On génère les abscisses du modèle de 0 à 1
                                                        
for i in X_modele_D:
    Y_modele_D = k_D * X_modele_D #On génère les ordonnées du modèle
```

### ETAPE 5 : On trace les trois graphiques avec les points expérimentaux et les modèles

### $\color {red} {"A\: VOUS\: DE\: JOUER\,!"}$ 
### Sur chaque graphique, comparer les modèles (droites bleues) à la répartition de vos points expérimentaux (points rouges).

## Modèle de Grosseteste

```python
plt.figure(figsize=(8,8)) #Définit la taille de l'affichage du graphique
plt.grid() # Affiche une grille sur le graphique
plt.title("Modèle de Grosseteste", size=22) #Définit le titre du graphique
plt.ylabel("angle d'incidence i (°)", size=18) #Affiche le titre des ordonnées
plt.xlabel("angle de réfraction r (°)", size=18) #Affiche le titre des abscisses
plt.ylim(0, 90) #définit le début et la fin des ordonnées
plt.xlim(0, 90) #Définit le début et la fin des abscisses.
plt.plot(angle_r, angle_i, 'o', color ='red', label = 'Vos points expérimentaux') #Affiche les points expérimentaux
plt.plot(X_modele_G, Y_modele_G, color = 'b', label = 'Modèle de Grosseteste') #Affiche le modèle mathématique, ici une droite affine
plt.legend(fontsize=18)# Affiche la légende
plt.text(40,50,"r = " + str(k_G) + "*" + "i", size=22, color='b', bbox=dict(boxstyle="round", ec=("k"), fc=("w"))) #Affiche l'équation de la droite dans un encadré
plt.show() #Affiche le graphique
```

## Modèle de Kepler

```python
plt.figure(figsize=(8,8)) #Définit la taille de l'affichage du graphique
plt.grid() # Affiche une grille sur le graphique
plt.title("Modèle de Kepler", size=22) #Définit le titre du graphique
plt.ylabel("angle d'incidence i (°)", size=18) #Affiche le titre des ordonnées
plt.xlabel("angle de réfraction r (°)", size=18) #Affiche le titre des abscisses
plt.ylim(0, 90) #définit le début et la fin des ordonnées
plt.xlim(0, 90) #Définit le début et la fin des abscisses.
plt.plot(angle_r, angle_i, 'o', color ='red', label = 'Vos points expérimentaux') #Affiche les points expérimentaux
plt.plot(X_modele_K, Y_modele_K, color = 'b', label = 'Modèle de Kepler') #Affiche le modèle mathématique, ici une droite affine
plt.legend(fontsize=18)# Affiche la légende
plt.text(40,50,"r = " + str(k_K) + "*" + "i", size=22, color='b', bbox=dict(boxstyle="round", ec=("k"), fc=("w"))) #Affiche l'équation de la droite dans un encadré
plt.show() #Affiche le graphique
```

## Modèle de Descartes

```python
plt.figure(figsize=(8,8)) #Définit la taille de l'affichage du graphique
plt.grid() # Affiche une grille sur le graphique
plt.title("Modèle de Descartes", size=22) #Définit le titre du graphique
plt.ylabel("sin(i)", size=18) #Affiche le titre des ordonnées
plt.xlabel("sin(r)", size=18) #Affiche le titre des abscisses
plt.ylim(0, 1) #définit le début et la fin des ordonnées
plt.xlim(0, 1) #Définit le début et la fin des abscisses.
plt.plot(sin_r, sin_i, 'o', color ='red', label = 'Vos points expérimentaux') #Affiche les points expérimentaux
plt.plot(X_modele_D, Y_modele_D, color = 'b', label = 'Modèle de Descartes') #Affiche le modèle mathématique, ici une droite affine
plt.legend(fontsize=18)# Affiche la légende
plt.text(0.5,0.5,"sin(i) = " + str(k_D) + "*" + "sin(r)", size=22, color='b', bbox=dict(boxstyle="round", ec=("k"), fc=("w"))) #Affiche l'équation de la droite dans un encadré
plt.show() #Affiche le graphique
```
